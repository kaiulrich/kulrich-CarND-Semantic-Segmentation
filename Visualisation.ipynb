## 2. Create epic mean loose graph  

```python
def readData(file):
    """
    Reads the epoch lost data from a csv file.
    """
    epochs = []
    with open(file) as dataFile:
        dataReader = csv.reader(dataFile)
        for row in dataReader:
            batch = []
            for value in row:
                batch.append(float(value))
            epochs.append(batch)
        
    return epochs
    
looses = readData('./results/1520108488.8096638/looses.csv')
```

```python
epics =[]
means = []

i = 1;
for row in looses:
    if(len(row) > 0):
        epics.append(i)
        means.append(numpy.mean(row))
        i = i + 1
```

```python
plt.plot(epics, means)
plt.title('Mean loos by Epic')
plt.ylabel('Mean loos')
plt.xlabel('Epic')

plt.show()
```
